```python
from zipfile import ZipFile
import shutil
zip_path = '/content/final_project_images.v1i.yolov8.zip'

with ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('/content/dataset')
!ls /content/dataset
```

```python
from ultralytics import YOLO
model = YOLO('yolov3.pt')

results = model.train( data='/content/dataset/data.yaml',
        epochs=50,
        imgsz=256,
        batch=16,
        name='meter_defect_detection',
        project='runs/detect',
        device=0,
        exist_ok=True,
        save=True,
        patience=20,
        optimizer='adam',
        lr0=0.001,
        momentum=0.937,
        weight_decay=0.0005,
        warmup_epochs=3.0,
        box=7.5,
        cls=0.5,
        dfl=1.5,
        augment=True,
        hsv_h=0.015,
        hsv_s=0.7,
        hsv_v=0.4,
        degrees=0.0,
        translate=0.1,
        scale=0.9,
        shear=0.0,
        perspective=0.0,
        flipud=0.0,
        fliplr=0.5,
        mosaic=1.0,
        mixup=0.15)
```
